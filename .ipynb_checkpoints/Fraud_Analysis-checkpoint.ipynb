```python
import pandas as pd

df = pd.read_csv("creditcard.csv")
print(df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
# The dataset contains 284,807 transactions (rows) and 31 columns (features + target):
# - Columns V1-V28: anonymized features from PCA transformation
# - Time: seconds elapsed since the first transaction
# - Amount: transaction amount
# - Class: target variable (0 = Legit, 1 = Fraud)
# This is a large dataset suitable for training robust models.
```

```python
#Create Time Features
import pandas as pd

# Assuming df is already loaded
# 1. Total hours since first transaction
df['Time_Hours'] = df['Time'] / 3600

# 2. Total days since first transaction
df['Time_Days'] = df['Time'] / (3600 * 24)

# 3. Hour of day (0-23) for daily pattern analysis
df['Hour_of_Day'] = (df['Time'] % 86400) // 3600
df['Hour_of_Day'] = df['Hour_of_Day'].astype(int)

# 4. Minute of hour (0-59) for finer pattern detection
df['Minute_of_Hour'] = ((df['Time'] % 3600) // 60).astype(int)

# 5. Second of minute (0-59) for complete temporal resolution
df['Second_of_Minute'] = (df['Time'] % 60).astype(int)

# Verify transformation
df[['Time', 'Time_Hours', 'Time_Days', 'Hour_of_Day', 'Minute_of_Hour', 'Second_of_Minute']].head()
```

```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Basic info
print("Dataset shape:", df.shape)
print("\nData types and missing values:")
print(df.info())
print("\nSummary statistics:")
print(df.describe())
print("\nClass distribution:")
print(df['Class'].value_counts())
```

```python
# -----------------------------
# Step: Handle Class Imbalance
# Reasoning:
# Fraud is extremely rare in this dataset (~0.17%). 
# If we train models on the raw dataset, they will almost always predict 'legit' 
# because that minimizes error. This will produce a useless model. 
# To make the model learn patterns in fraud, we balance the dataset using SMOTE.
# SMOTE (Synthetic Minority Oversampling Technique) generates synthetic samples 
# of the minority class (fraud) instead of just duplicating existing rows.
# This helps models generalize and detect fraud better.
# -----------------------------

from imblearn.over_sampling import SMOTE
import pandas as pd

# Separate features (X) and target (y)
# X: all columns except 'Class'
# y: 'Class' column (0=Legit, 1=Fraud)
X = df.drop(['Class'], axis=1)
y = df['Class']

# Initialize SMOTE with a fixed random state for reproducibility
smote = SMOTE(random_state=42)

# Apply SMOTE to create a balanced dataset
# X_resampled: new features including synthetic fraud samples
# y_resampled: new target array with balanced classes
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check and compare class distribution before and after resampling
print("Original dataset class distribution:")
print(y.value_counts())
print("\nResampled dataset class distribution:")
print(pd.Series(y_resampled).value_counts())

# Reasoning:
# Now we have a roughly equal number of fraud and legit transactions.
# This is crucial for training models that can detect fraud accurately.
```

```python
# -----------------------------
# Step: Handle Class Imbalance with SMOTE
# -----------------------------
# Original dataset class distribution:
# - Legit (0): 284,315 transactions
# - Fraud (1): 492 transactions (~0.17% of total)
# 
# Issue:
# - The dataset is extremely imbalanced.
# - Training models on raw data would lead them to predict mostly 'Legit'.
# - This would produce a model with high accuracy but poor fraud detection (useless in banking).
#
# Solution:
# - SMOTE (Synthetic Minority Oversampling Technique) is applied to generate synthetic
#   fraud transactions based on existing patterns.
# - Balances the dataset without simply duplicating rows.
#
# Resampled dataset class distribution:
# - Legit (0): 284,315
# - Fraud (1): 284,315
#
# Outcome:
# - Now both classes have equal representation.
# - The model can learn meaningful patterns for both fraud and legit transactions.
# - Essential for building a robust fraud detection model suitable for real-world banking applications.
```

```python
import numpy as np

# -----------------------------
# Step: Feature Engineering for Fraud Detection
# Reasoning:
# Creating derived features helps the model capture patterns not obvious in raw data.
# This is especially important in fraud detection where fraud often occurs at unusual
# times or amounts.
# -----------------------------

# 1. Amount-based transformations
df_resampled = X_resampled.copy()  # start from balanced dataset
df_resampled['Amount_log'] = np.log1p(df_resampled['Amount'])  # Log transform to reduce extreme outliers
df_resampled['Amount_squared'] = df_resampled['Amount'] ** 2   # Highlight unusually large transactions
df_resampled['Amount_ratio'] = df_resampled['Amount'] / (df_resampled['Amount'].mean())  # Relative to average

# 2. Time-based features
df_resampled['Is_Night'] = df_resampled['Hour_of_Day'].apply(lambda x: 1 if x >=0 and x <=6 else 0)  # Flag night transactions

# 3. Interaction features
df_resampled['Amount_x_Hour'] = df_resampled['Amount'] * df_resampled['Hour_of_Day']  # Large transactions at unusual hours
df_resampled['Amount_per_Hour'] = df_resampled['Amount'] / (df_resampled['Time_Hours'] + 1)  # Transaction velocity (+1 to avoid divide by 0)

# -----------------------------
# Summary:
# - These new features give the model more signals to distinguish fraud.
# - Amount features capture unusual transaction sizes.
# - Time features capture unusual transaction times.
# - Interaction features combine amount and time for complex patterns.
# -----------------------------

# Combine with target variable
df_resampled['Class'] = y_resampled
df_resampled.head()

```

```python
from sklearn.model_selection import train_test_split

# -----------------------------
# Step: Train/Test Split
# Reasoning:
# Splitting the dataset ensures that the model is evaluated on data it has not seen before.
# Stratifying by 'Class' keeps the fraud/legit ratio equal in both train and test sets.
# -----------------------------

# Separate features and target
X_final = df_resampled.drop('Class', axis=1)
y_final = df_resampled['Class']

# Perform stratified split: 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_final, test_size=0.2, random_state=42, stratify=y_final
)

# Check shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train distribution:\n", y_train.value_counts())
print("y_test distribution:\n", y_test.value_counts())

```

```python
# -----------------------------
# Step: Train/Test Split
# -----------------------------
# After resampling and feature engineering, we split the dataset into training and testing sets:
# - X_train shape: (454,904, 41) → 454,904 samples, 41 features
# - X_test shape:  (113,726, 41) → 113,726 samples reserved for testing
#
# y_train distribution:
# - Legit (0): 227,452 transactions
# - Fraud (1): 227,452 transactions
#
# y_test distribution:
# - Legit (0): 56,863 transactions
# - Fraud (1): 56,863 transactions
#
# Interpretation:
# - The split preserves **class balance** in both training and testing sets.
# - Ensures that models are trained on a balanced dataset and evaluated on a balanced, representative test set.
# - X_train contains enough samples for the model to learn complex patterns.
# - X_test allows for accurate, unbiased assessment of model performance.
# - Random shuffling ensures that the training and test sets are representative of the overall dataset.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# Step: Baseline Logistic Regression
# Reasoning:
# Logistic Regression is simple and interpretable, giving us a first benchmark.
# Outputs probabilities, useful for ROC-AUC, and allows us to understand feature importance.
# -----------------------------

# Initialize model
lr = LogisticRegression(max_iter=1000, random_state=42)

# Train the model on the training set
lr.fit(X_train, y_train)

# Predict class labels on test set
y_pred = lr.predict(X_test)

# Predict probabilities for ROC-AUC calculation
y_prob = lr.predict_proba(X_test)[:,1]

# -----------------------------
# Step: Evaluation
# We focus on:
# - Precision: How many predicted frauds are actually fraud
# - Recall: How many actual frauds were detected
# - F1-score: Balance between precision and recall
# - ROC-AUC: Overall discrimination ability
# -----------------------------

# 1. Classification report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

# 2. Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Legit','Fraud'], yticklabels=['Legit','Fraud'])
plt.title("Confusion Matrix: Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# 3. ROC-AUC score
roc_score = roc_auc_score(y_test, y_prob)
print("ROC-AUC Score:", round(roc_score, 4))

# 4. ROC Curve plot for visualization
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, color='red', label=f'ROC Curve (AUC = {roc_score:.4f})')
plt.plot([0,1], [0,1], color='blue', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve: Logistic Regression")
plt.legend()
plt.show()

```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -----------------------------
# Step: Model Evaluation and Visualization
# -----------------------------

# 1. Classification Report (text output)
print("\n=== Classification Report ===\n")
print(classification_report(y_test, y_pred))

# 2. Confusion Matrix (text output)
cm = confusion_matrix(y_test, y_pred)
print("\n=== Confusion Matrix ===\n")
print(cm)

# 3. Confusion Matrix (visual)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Legit','Fraud'], yticklabels=['Legit','Fraud'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# 4. ROC Curve and AUC
roc_score = roc_auc_score(y_test, y_prob)
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, color='red', label=f'ROC Curve (AUC = {roc_score:.4f})')
plt.plot([0,1], [0,1], color='blue', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
print(f"\nROC-AUC Score: {roc_score:.4f}")

# 5. Precision-Recall Curve (optional but informative)
precision, recall, pr_thresholds = precision_recall_curve(y_test, y_prob)
avg_precision = average_precision_score(y_test, y_prob)
plt.figure(figsize=(8,6))
plt.plot(recall, precision, color='green', label=f'Precision-Recall Curve (AP = {avg_precision:.4f})')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
print(f"\nAverage Precision Score: {avg_precision:.4f}")
```

```python
# -----------------------------
# Logistic Regression – Evaluation Summary
# -----------------------------
# This output shows both the classification report and confusion matrix,
# providing a detailed view of model performance on the test dataset.

# Classification Report Interpretation:
# - Accuracy: 97%
#   → The model correctly classifies 97% of all transactions.

# Class 0 (Legitimate Transactions):
# - Precision = 0.96
#   → 96% of transactions predicted as legitimate are actually legitimate.
# - Recall = 0.98
#   → The model correctly identifies 98% of all legitimate transactions.
# - F1-score = 0.97
#   → Strong overall performance for legitimate transactions.

# Class 1 (Fraudulent Transactions):
# - Precision = 0.98
#   → When the model flags a transaction as fraud, it is correct 98% of the time.
# - Recall = 0.95
#   → The model detects 95% of all fraud cases.
# - F1-score = 0.97
#   → Indicates strong fraud detection capability, though some fraud is still missed.

# -----------------------------
# Confusion Matrix Interpretation:
# -----------------------------
# [[55762  1101]
#  [ 2602 54261]]

# Breakdown:
# - True Negatives (TN) = 55,762
#   → Legitimate transactions correctly classified as legit.
# - False Positives (FP) = 1,101
#   → Legitimate transactions incorrectly flagged as fraud.
# - False Negatives (FN) = 2,602
#   → Fraudulent transactions missed by the model (high business risk).
# - True Positives (TP) = 54,261
#   → Fraudulent transactions correctly identified.

# Business Interpretation:
# - The model performs very well overall but still misses 2,602 fraud cases.
# - In a banking context, false negatives are costly because missed fraud
#   leads to direct financial losses.
# - While Logistic Regression is interpretable and strong as a baseline,
#   this level of missed fraud justifies using more advanced models
#   (e.g., Random Forest and XGBoost) to improve recall for fraud detection.

# Conclusion:
# - Logistic Regression serves as a strong baseline model.
# - Further improvements are necessary to minimize missed fraud cases,
#   which is critical for real-world banking applications.
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -----------------------------
# Step: Random Forest Classifier
# Reasoning:
# Random Forest handles large datasets efficiently, captures non-linear relationships,
# reduces overfitting, and provides feature importance.
# -----------------------------

# Initialize Random Forest
rf = RandomForestClassifier(
    n_estimators=200,       # number of trees
    max_depth=15,           # limit depth to prevent overfitting
    random_state=42,
    n_jobs=-1               # use all CPU cores
)

# Train the model
rf.fit(X_train, y_train)

# Predict on test set
y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]  # probabilities for ROC-AUC

# -----------------------------
# Step: Model Evaluation
# -----------------------------

# Classification report
print("\n=== Classification Report: Random Forest ===\n")
print(classification_report(y_test, y_pred_rf))

# Confusion matrix (text)
cm_rf = confusion_matrix(y_test, y_pred_rf)
print("\n=== Confusion Matrix ===\n")
print(cm_rf)

# Confusion matrix (visual)
plt.figure(figsize=(6,5))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', xticklabels=['Legit','Fraud'], yticklabels=['Legit','Fraud'])
plt.title("Confusion Matrix: Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC curve and AUC
roc_score_rf = roc_auc_score(y_test, y_prob_rf)
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_prob_rf)
plt.figure(figsize=(8,6))
plt.plot(fpr_rf, tpr_rf, color='red', label=f'ROC Curve (AUC = {roc_score_rf:.4f})')
plt.plot([0,1], [0,1], color='blue', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve: Random Forest")
plt.legend()
plt.show()
print(f"\nROC-AUC Score: {roc_score_rf:.4f}")

# -----------------------------
# Step: Feature Importance
# Reasoning:
# Shows which features contributed most to fraud detection,
# very useful for explainability in banking projects.
# -----------------------------

importances = rf.feature_importances_
feature_names = X_train.columns

# Sort and visualize top 15 features
indices = np.argsort(importances)[::-1][:15]
plt.figure(figsize=(10,6))
sns.barplot(x=importances[indices], y=feature_names[indices], palette="viridis")
plt.title("Top 15 Feature Importances: Random Forest")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()
```

```python
# -----------------------------
# Fraud Detection: XGBoost + SHAP
# -----------------------------

import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import shap
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -----------------------------
# 1. Initialize XGBoost
# Reasoning:
# XGBoost is a high-performance gradient boosting model,
# handles non-linear relationships, large datasets efficiently,
# and allows feature impact explanation via SHAP.
# -----------------------------
xgb_model = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=8,
    learning_rate=0.1,
    scale_pos_weight=1,   # dataset is already balanced
    random_state=42,
    n_jobs=-1,
    use_label_encoder=False,
    eval_metric='logloss'
)

# -----------------------------
# 2. Train the model
# -----------------------------
xgb_model.fit(X_train, y_train)

# -----------------------------
# 3. Predict on test set
# -----------------------------
y_pred_xgb = xgb_model.predict(X_test)
y_prob_xgb = xgb_model.predict_proba(X_test)[:,1]

# -----------------------------
# 4. Evaluation: Classification Report & Confusion Matrix
# -----------------------------
print("\n=== Classification Report: XGBoost ===\n")
print(classification_report(y_test, y_pred_xgb))

cm_xgb = confusion_matrix(y_test, y_pred_xgb)
print("\n=== Confusion Matrix ===\n")
print(cm_xgb)

plt.figure(figsize=(6,5))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues', xticklabels=['Legit','Fraud'], yticklabels=['Legit','Fraud'])
plt.title("Confusion Matrix: XGBoost")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# -----------------------------
# 5. ROC Curve & AUC
# -----------------------------
roc_score_xgb = roc_auc_score(y_test, y_prob_xgb)
fpr_xgb, tpr_xgb, thresholds_xgb = roc_curve(y_test, y_prob_xgb)
plt.figure(figsize=(8,6))
plt.plot(fpr_xgb, tpr_xgb, color='red', label=f'ROC Curve (AUC = {roc_score_xgb:.4f})')
plt.plot([0,1], [0,1], color='blue', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve: XGBoost")
plt.legend()
plt.show()
print(f"\nROC-AUC Score: {roc_score_xgb:.4f}")

# -----------------------------
# 6. SHAP Analysis: Feature Importance
# Reasoning:
# SHAP explains how each feature contributes to predictions,
# very valuable for stakeholders in banking fraud detection.
# -----------------------------
explainer = shap.Explainer(xgb_model, X_train)
shap_values = explainer(X_test)

# Summary plot of top 15 features
shap.summary_plot(shap_values, X_test, plot_type="bar", max_display=15, show=True)
```
